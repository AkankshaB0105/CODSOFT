```python
with open('/content/drive/MyDrive/codsoftdata/description.txt','r', encoding='utf-8') as file:
  descriptions = file.readlines()
```

```python
with open('/content/drive/MyDrive/codsoftdata/train_data.txt','r', encoding='utf-8') as file:
  genres = file.readlines()
```

```python
# ENSURE THAT LENGTHS ARE SAME:
min_length = min(len(descriptions),len(genres))
descriptions = descriptions[:min_length]
genres = genres[:min_length]
```

```python
#CREATE A DATAFRAME
train_data = pd.DataFrame({'Description': descriptions, 'Genre': genres})
```

```python
#PREPROCESS THE DATA
train_data['Description'] = train_data['Description'].str.lower().str.strip()
train_data['Genre'] = train_data['Genre'].str.strip()
```

```python
#PRINT SOME INFORMATION FOR DEBUGGING
print("Data head:")
print(train_data.head())
```

```python
#TEXT VECTORIZATION
vectorizer = TfidfVectorizer(max_features=5000, stop_words='english')
x = vectorizer.fit_transform(train_data['Description'])
y = train_data['Genre']
```

```python
#SPLIT THE DATA INTO TRAINING AND TESTING SETS
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)
```

```python
#TRAIN A MULTINOMIAL NAIVE BAYES CLASSIFIER
classifier = MultinomialNB()
classifier.fit(x_train, y_train)
```

```python
#MAKE PREDICTIONS ON THE TEST SET
y_pred = classifier.predict(x_test)
```

```python
# DISPLAY THE DISTRIBUTION OF GENRES IN THE DATASET
genre_distribution = train_data['Genre'].value_counts()
print("Classification Report: ")
print(genre_distribution)
```
